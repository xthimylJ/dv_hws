```python
import pandas as pd
```

#### Exercise 1: Student Grades with Highest Score

```python
table = {"Names": ["Ahmed", "Muhtar", "Emirhan", "Nazima", "Aliya"], 
         "Math": [98, 57, 66, 75, 82],
         "English": [91, 58, 78, 23, 56],
         "History": [56, 89, 58, 63, 77],
         "OOP": [88, 96, 45, 75, 36]
         }

df = pd.DataFrame(table)
df["Highest Score"] = df.iloc[:, 1:].max(axis=1)

df
```

#### Exercise 2: Class Performance Summary


```python
table2 = {"Names": ["Ahmed", "Muhtar", "Emirhan", "Nazima"], 
         "Math": [98, 57, 66, 75],
         "English": [91, 58, 78, 23],
         "History": [56, 89, 58, 63],
         
         }

df2 = pd.DataFrame(table2)
df2["Mean Score"] = df.iloc[:, 1:].mean(axis=1)
df2["Pass/Fail"] = df2["Mean Score"].apply(lambda x: "Pass" if x >= 50 else "Fail") 
df2
```

#### Exercise 3: Course-Wise Averages

```python
table3 = {"Names": ["Ahmed", "Muhtar", "Emirhan", "Nazima", "Charlie", "Bob"], 
         "Math": [98, 57, 66, 75, 78, 99],
         "English": [91, 58, 78, 23, 85, 45],
         "History": [56, 89, 58, 63, 77, 64],
         
         }

df3 = pd.DataFrame(table3)
df3.loc["Course Average"] = df.mean(numeric_only=True)
df3
```

#### Exercise 4: Ranking Students


```python
table4 = {"Names": ["Ahmed", "Muhtar", "Emirhan", "Nazima", "Aliya"], 
         "Math": [98, 57, 66, 75, 82],
         "English": [91, 58, 78, 23, 56],
         "History": [56, 89, 58, 63, 77],
         "OOP": [88, 96, 45, 75, 36]
         }

df4 = pd.DataFrame(table4)

df4["Total marks"] = df.iloc[:, 1:].sum(axis=1)
df4_sorted = df4.sort_values(by="Total marks", ascending=False)
df4_sorted["Rank"] = df4_sorted["Total marks"].rank(method="dense", ascending=False).astype(int)
df4_sorted
```

#### Exercise 5: Adding a Bonus Mark

```python
students = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Math": [45, 78, 92, 55],
    "Science": [50, 82, 87, 60],
    "English": [48, 80, 85, 58]
}

df5 = pd.DataFrame(students)
df5.iloc[:, 1:] = df5.iloc[:, 1:].map(lambda x: min(x + 5, 100))
df5["New Average Score"] = df5.iloc[:, 1:].mean(axis=1)
df5
```
